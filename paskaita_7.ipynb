```python
#kur ilgiausi aprasymai
wine_df['description_len'] = wine_df['description'].map(len)
wine_df.head()
```

```python
def word_count(txt):
    return len(txt.split())
```

```python
wine_df['description_words'] = wine_df['description'].map(word_count)
wine_df.head()
```

```python
#norime sujungti du stulpelius i viena
def my_function(row):
    return str(row['country']) + ' ' + str(row['province'])
```

```python
wine_df['place'] = wine_df.apply(my_function, axis=1) #axis=1 nurodom, kad gauti informacija apie visa eilute (o ne stulpeli =0)
wine_df
```

# DROP NaNs

```python
wine_df_cleaned = wine_df.drop_duplicates() # jeigu nieko neirasom i skliaustus, paima pirma unikalia reiksme, o kitas ismeta
wine_df_cleaned
```

```python
wine_df_cleaned[wine_df_cleaned['price'].isnull()].shape #shape duoda rezulto forma
```

```python
#perrasom tikrus duomenis
wine_df_cleaned = wine_df_cleaned.dropna(how='any', subset=['price', 'province']) # drop rows with NaN in price, province collums
```

# GROUP BY

```python
# sugrupuoti duomenis pagal kazkoki faktoriu

grouped = wine_df_cleaned.groupby(['country', 'province'])[['points','price']].mean() #grupuoti pagal vidurki
```
