# JOINTLOT

```python
sns.jointplot(data=tips, x='total_bill', y='tip')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind="reg") #reg - regresion. LINIJA BANDO PROGNOZUOTI. KIEKVIENAS TASKAS VAIZDUOJA NAUJA EILUTE LENTELEJE
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind="hex") #kuo tamsesnis, tuo daugiau ten yra reiksmiu 
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue="day") 
```

# PAIRPLOT

```python
# automatiskai palygina galimus stulpelius
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex',palette='magma') #galima tik pagal viena aspekta 
```

```python
# NUSIRASYTI KAS DAR TEN LIKO PO SITO. SU DVIEM SPALVOM
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', size='size') #SCATTERPLOT
```

```python
ax = sns.scatterplot(data=tips, x='total_bill', y='tip', size='size', hue='day') #kad legenda butu sone
ax.legend()
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
ax = sns.scatterplot(data=tips, x='total_bill', y='tip', size='size', hue='day', style='time')
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
sns.regplot(data=tips, x='total_bill', y='tip', order=1) #rodo regresija order1 - LINIJINE. Duomenys tiesiskai priklauso nuo skaiciaus pvz 1
```

```python
sns.regplot(data=tips, x='total_bill', y='tip', order=2) #POLYNOMIAL REGRESSION. duomenys NERA tiesiskai priklauso nuo skaiciaus pvz 1
```

# HEATMAP

```python
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
flights_pivot = flights.pivot('month','year','passengers')
flights_pivot
```

```python
sns.heatmap(flights_pivot)
```

```python
sns.heatmap(flights_pivot, annot=True, fmt='d', linewidths=0.5)
```
